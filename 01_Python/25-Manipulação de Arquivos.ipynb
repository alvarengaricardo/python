```python
try:
    manipulador = open('texto.txt', 'r', encoding = 'utf-8')
    for linha in manipulador:
        print(linha, end = "")
except IOError:
    print('Não foi possível abrir o arquivo.')
else:
    manipulador.close()
```

```python
# Procurando texto no arquivo
texto = 'urna'
try:
    manipulador = open('texto.txt', 'r', encoding = 'utf-8')
    for linha in manipulador:
        if texto in linha:
            print(linha, end = "")
except IOError:
    print('Não foi possível abrir o arquivo.')
else:
    manipulador.close()
```

```python
try:
    manipulador = open('saida', 'w', encoding = 'utf-8')
    for linha in manipulador:
        print(linha, end = "")
except IOError:
    print('Não foi possível abrir o arquivo.')
else:
    manipulador.close()
```

```python
try:
    manipulador = open('saida.txt', 'w+', encoding = 'utf-8')
    for numero in range(21):
        linha = str(numero) + '\n'
        manipulador.write(linha)
except IOError:
    print('Não foi possível abrir o arquivo.')
else:
    manipulador.close()
```

```python
try:
    manipulador = open('saida.txt', 'r', encoding = 'utf-8')
    for linha in manipulador:
        print(linha, end ="")
except IOError:
    print('Não foi possível abrir o arquivo.')
else:
    manipulador.close()
```

```python
frutas = ['Morango\n', 'Abacate\n', 'Uva\n', 'Abacaxi\n', 'Banana']
try:
    manipulador = open('frutas.txt', 'w', encoding = 'utf-8')
    manipulador.writelines(frutas)
except IOError:
    print('Não foi possível abrir o arquivo.')
else:
    manipulador.close()
```

```python
try:
    manipulador = open('frutas.txt', 'r', encoding = 'utf-8')
    print(manipulador.read())
except IOError:
    print('Não foi possível abrir o arquivo.')
else:
    manipulador.close()
```
