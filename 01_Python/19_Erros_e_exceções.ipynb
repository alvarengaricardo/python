# Erros, Exceção e Raise

##### Exceção é um objeto que representa um erro que ocorreu ao executar o programa

Usa-se os blocos try... except

```python
# tratando erro divisão por zero
```

```python
n1 = int(input('n1: '))
```

```python
n2 = int(input('n2: '))
```

```python
try:
  r = round(n1 / n2, 2)
except ZeroDivisionError:
  print('Divisão por zero')
else:
  print(r)
finally:
  print('Fim')
```

# Raise

```python
from math import sqrt
```

```python
class num_negativo_Error(Exception):
  def __init__(self):
    pass
```

```python
try:
  num = int(input('número positivo: '))
  if num < 0:
    raise ArithmeticError
except ArithmeticError:
  print('Número negativo!')
else:
  print(sqrt(num))
finally:
  print('Fim')
```

```python
try:
  num = int(input('número positivo: '))
  if num < 0:
    raise num_negativo_Error
except num_negativo_Error:
  print('Número menor que zero!')
else:
  print(sqrt(num))
finally:
  print('Fim')
```
