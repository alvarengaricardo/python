# Operadores LÃ³gicos - AND, OR, NOT

```python
def valida(idade, altura):
    return (idade >= 18) and (altura >= 1.7)
```

```python
valida(25, 1.75)
```

```python
valida(17, 1.75)
```

```python
def alarme(janela, porta):
    return (porta == 'aberta') or (janela == 'aberta')
```

```python
alarme('fechada', 'fechada')
```

```python
alarme('aberta', 'fechada')
```
